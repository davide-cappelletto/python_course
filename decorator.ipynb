```python
class Movie:

    def __init__(self, title, rating):
        self.title = title
        self._rating = rating
    
    @property
    def rating(self):
        print("Calling the getter...")
        return self._rating
    
    @rating.setter
    def rating(self, new_rating):
        print("Calling the setter...")
        if isinstance(new_rating, float) and 1.0 <= new_rating <= 5.0:
            self._rating = new_rating
        else:
            print("Please enter a valid rating.")



favorite_movie = Movie("Titanic", 4.3)
print(favorite_movie.rating)

favorite_movie.rating = 4.5
print(favorite_movie.rating)
```

```python
class Backpack:

    def __init__(self):
        self._items = []
    
    @property
    def items(self):
        print("Calling the getter...")
        return self._items
    
    @items.setter
    def items(self, new_items):
        print("Calling the setter...")
        if isinstance(new_items, list):
            self._items = new_items
        else:
            print("Please enter a valid list of items.")


my_backpack = Backpack()

print(my_backpack.items)
my_backpack.items = ["Water Bottle", "Sleeping Bag"]
print(my_backpack.items)
```

```python
class BouncyBall:
    
	def __init__(self, price, size, brand):
		self._price = price
		self._size = size
		self._brand = brand
	
	@property
	def price(self):
		print("Calling the 'price' getter...")
		return self._price
	
	@price.setter
	def price(self, new_price):
		print("Calling the 'price' setter...")
		if isinstance(new_price, float):
			self._price = new_price
		else:
			print("Please enter a valid price.")
	
	@property
	def size(self):
		print("Calling the 'size' getter...")
		return self._size
	
	@size.setter
	def size(self, new_size):
		print("Calling the 'size' setter...")
		if isinstance(new_size, str):
			self._size = new_size
		else:
			print("Please enter a valid size.")

	@property
	def brand(self):
		print("Calling the 'brand' getter...")
		return self._brand
	
	@brand.setter
	def brand(self, new_brand):
		print("Calling the 'brand' setter...")
		if isinstance(new_brand, str):
			self._brand = new_brand
		else:
			print("Please enter a valid brand.")


my_ball = BouncyBall(7.99, "Small", "ToysRFun")
print(my_ball.price)
print(my_ball.size)
print(my_ball.brand)

my_ball.price = 10.99
my_ball.size = "Medium"
my_ball.brand = "PlayW/You"

print(my_ball.price)
print(my_ball.size)
print(my_ball.brand)
```
