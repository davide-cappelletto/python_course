```python
class CashRegister:

    def __init__(self, name):

        self._purchase = {'name':[], 'price':[]}
        self._cashier_name = name
    
    @property
    def purchase(self):
        return self._purchase
    
    @property
    def cashier_name(self):
        return self._cashier_name
    
    #Methods
    
    def add_products(self, product, price, quantity=1):

        if isinstance(product, str) and isinstance(price, float):
            q = 1
            while q <= quantity:
                self.purchase['name'].append(product)
                self.purchase['price'].append(price)
                q += 1
        else:
            print("Please provide a valid product.")
    
    def show_products(self):

        print(self.purchase)
    
    def remove_product(self, product):

        if product in self.purchase['name']:
            i = self.purchase['name'].index(product)
            del self.purchase['name'][i], self.purchase['price'][i]
        else:
            print('This product is not listed.')

    
    def update_price(self, product, new_price):

        for index, value in enumerate(self.purchase['name']):
            if value == product:
                self.purchase['price'][index] = new_price
    
    def subtotal(self):
        
        self.sub = sum(self.purchase['price'])
        print(f"Subtotal: {self.sub}")

    def taxes(self):
        
        self.tax = self.sub * 0.05
        print(f"Taxes: {self.tax}")
    
    def total(self):

        self.tot = self.sub + self.tax
        print(f"Total: {self.tot}")

    def clear_products(self):
        self.purchase['name'] = []
        self.purchase['price'] = []
        print('Purchase has been cleared.')


    
```

```python
frank = CashRegister('Frank')
frank.add_products('Milk', 1.99, 2)
frank.add_products('Eggs', 2.49, 1)
frank.add_products('Flour', 1.29, 1)
frank.show_products()
frank.remove_product('Milk')
frank.show_products()
frank.remove_product('Yeast')
frank.update_price('Milk',2.19)
frank.show_products()
frank.subtotal()
frank.taxes()
frank.total()
frank.clear_products()
frank.show_products()
```
